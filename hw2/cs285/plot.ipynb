```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
theta = np.linspace(0.0001, 0.9999, 100000)  # Avoid division by zero by starting from 0.01 and ending at 0.99
y1 = (4 * theta**2 + 8 * theta + 1) / ((1 - theta)**4 * theta)
y2 = (theta**2 + 3 * theta + 1) / ((1 - theta)**4 * theta)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(theta, y1, label=r'$y=\frac{4\theta^{2}+8\theta+1}{(1-\theta)^{4}\theta}$')
plt.plot(theta, y2, label=r'$y=\frac{\theta^{2}+3\theta+1}{(1-\theta)^{4}\theta}$')
plt.title("Plot of the given functions")
plt.xlabel(r"$\theta$")
plt.ylabel("y")
plt.legend()
plt.ylim(15, 500)  # Set y-axis limits
plt.grid(True)
plt.tight_layout()
plt.show()
```
